About Dataset

"Imagine a customer placing an order on Blinkit, expecting it to arrive within minutes. Behind the scenes, multiple things happen — the order is assigned to a store, packed, handed over to a delivery partner, and finally delivered to the customer. This dataset captures that entire journey. It records when the order was placed, when it was promised to be delivered, and when it actually arrived. By analyzing this data, we can understand where delays happen, which delivery partners or stores perform better, and how different factors like payment methods influence delivery speed. Essentially, this dataset helps us uncover insights that can improve delivery efficiency and enhance the overall customer experience."
