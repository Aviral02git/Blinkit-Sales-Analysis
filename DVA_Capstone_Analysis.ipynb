## **About the Dataset**
# Overview
This dataset contains comprehensive, cleaned data representing an e-commerce or quick-commerce retail environment. It provides a 360-degree view of the business operations, combining transactional records, product inventory details, delivery logistics, and customer feedback into a single, unified view suitable for dashboarding and advanced analytics.

Key Data Domains Included:

🛒 Order & Transaction Details: * order_id, order_date, order_total, payment_method, quantity, unit_price, total_price.

Time-based features for trend analysis (hour, day, month, weekday).

👥 Customer Profiling: * customer_id, customer_segment, registration_date, total_orders, avg_order_value.

Geographic data including area and pincode.

📦 Product & Inventory Metrics: * product_id, product_name, category (e.g., Pet Care, Instant Food), brand.

Financial and operational metrics: mrp, margin_percentage, shelf_life_days, min_stock_level, max_stock_level.

🚚 Logistics & Delivery Performance:

delivery_partner_id, store_id, distance_km.

Time tracking: promised_time, actual_time, delivery_time_minutes.

Delay tracking: is_delayed, delay_minutes, reasons_if_delayed, delay_bucket, delay_category, delivery_status.

⭐ Customer Feedback & Sentiment:

feedback_id, rating, feedback_text.

Categorized feedback metrics: feedback_category, sentiment (e.g., Neutral, Positive, Negative), feedback_date.

📈 Derived Analytical Features:

Pre-calculated metrics for easier analysis, such as delivery_efficiency, order_size, is_peak (peak hour flag), and order_value_bucket.

Potential Use Cases:

Delivery Optimization: Predicting delays and analyzing root causes of late deliveries based on distance, peak times, and partner IDs.

Customer Sentiment Analysis: Correlating customer ratings and NLP sentiment with delivery delays and product categories.

Sales & Inventory Forecasting: Tracking high-demand items, monitoring margin percentages, and optimizing stock levels across different areas.

## **Datetime Conversion**

The notebook plots monthly trends and hourly delays, but it relies entirely on the pre-extracted month and hour integer columns. The actual order_date, promised_time, and actual_time columns are still formatted as plain text (strings).

It needs fixing because if we want to do any advanced time-series analysis later (e.g., plotting day-by-day continuous revenue, or accurately calculating the exact minutes between promised_time and actual_time across midnight boundaries), Pandas will fail because it cannot do math on text.

```python
# Convert all time-based columns to Datetime objects for future-proofing
date_cols = ['order_date', 'promised_time', 'actual_time', 'feedback_date', 'registration_date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], errors='coerce')

print("Dates converted successfully. Data types are now:")
print(df[date_cols].dtypes)
```

**Revenue by Category**

```python
import matplotlib.pyplot as plt
import seaborn as sns

cat_revenue = df.groupby('category').agg(
    revenue      = ('total_price', 'sum'),
    orders       = ('order_id', 'nunique'),
    avg_margin   = ('margin_percentage', 'mean')
).reset_index().sort_values('revenue', ascending=False)

# Bar chart with a solid, highly visible professional blue
plt.figure(figsize=(12, 6))
sns.barplot(data=cat_revenue, x='revenue', y='category', color='#1F3864')
plt.title('Total Revenue by Product Category', fontsize=14, fontweight='bold')
plt.xlabel('Revenue')
plt.ylabel('')
plt.tight_layout()
plt.savefig('category_revenue_bar.png', dpi=150)
plt.show()
```

**Delay by Hour (Peak vs Off-Peak)**

```python
import matplotlib.pyplot as plt
import seaborn as sns

hourly_delay = df.groupby('hour')['delay_minutes'].mean().reset_index()

plt.figure(figsize=(14,5))
sns.barplot(data=hourly_delay, x='hour', y='delay_minutes', palette='Blues_d')
plt.title('Average Delay by Hour of Day')
plt.axvspan(7.5, 11.5, alpha=0.1, color='red', label='Morning Peak')
plt.axvspan(17.5, 22.5, alpha=0.1, color='orange', label='Evening Peak')
plt.legend()
plt.savefig('delay_by_hour.png', dpi=150)
plt.show()
```

**Store Performance**

**Peak vs Off-Peak Comparison**

```python
peak_compare = df.groupby('is_peak').agg(
    avg_delay        = ('delay_minutes', 'mean'),
    avg_efficiency   = ('delivery_efficiency', 'mean'),
    avg_actual_time  = ('actual_duration_minutes', 'mean'),
    order_count      = ('order_id', 'count')
).reset_index()

peak_compare['is_peak'] = peak_compare['is_peak'].map({True:'Peak', False:'Off-Peak'})
print(peak_compare)
```

**Orders by Day/Month Trend**

```python
import matplotlib.pyplot as plt

# 1. Create a proper chronological Year-Month column (e.g., '2023-03')
df['year_month'] = df['order_date'].dt.strftime('%Y-%m')

# 2. Group by this new linear timeline
monthly_orders = df.groupby('year_month').agg(
    orders  = ('order_id', 'nunique'),
    revenue = ('order_total', 'sum')
).reset_index()

# Sort chronologically just to be safe
monthly_orders = monthly_orders.sort_values('year_month')

# 3. Plot the fixed trendline
fig, ax1 = plt.subplots(figsize=(14, 6))
ax2 = ax1.twinx()

# Bar chart for Orders
ax1.bar(monthly_orders['year_month'], monthly_orders['orders'], color='#1F3864', alpha=0.7, label='Orders')
# Line chart for Revenue
ax2.plot(monthly_orders['year_month'], monthly_orders['revenue'], color='#D6E4F0', linewidth=3, marker='o', label='Revenue')

plt.title('True Chronological Monthly Orders vs Revenue', fontsize=14, fontweight='bold')
ax1.set_xlabel('Month')
ax1.set_ylabel('Total Orders', color='#1F3864')
ax2.set_ylabel('Total Revenue (Cart Value)', color='#D6E4F0')

# Rotate the x-axis labels so the Year-Month text doesn't overlap
ax1.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.savefig('true_monthly_trends.png', dpi=150)
plt.show()
```

## **Dashboard 4 — Inventory & Profitability (Supply Chain)**
**Key questions this dashboard answers:**
1) Which product categories are driving the highest profit margins?
2) What is the shelf-life risk across our inventory?
3) Are we maintaining healthy stock levels (Min vs. Max) for our most profitable items?
4) Is there a relationship between how quickly an item expires and how much profit it makes?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Supply Chain KPIs
print("--- SUPPLY CHAIN KPIs ---")
print(f"Average Profit Margin: {df['margin_percentage'].mean():.2f}%")
print(f"Average Shelf Life: {df['shelf_life_days'].mean():.0f} days")
print(f"Total Unique Products: {df['product_id'].nunique()}")
print("-" * 25, "\n")
```

**Chart 1: Average Margin by Category**

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
margin_cat = df.groupby('category')['margin_percentage'].mean().sort_values(ascending=False).reset_index()

# The line break is fixed below
sns.barplot(data=margin_cat, x='margin_percentage', y='category', color='forestgreen')
plt.title('1. Average Profit Margin by Category', fontsize=14, fontweight='bold')

plt.xlabel('Margin %')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

**Chart 2: Shelf Life Risk by Category**

```python
plt.figure(figsize=(10, 6))
shelf_life_cat = df.groupby('category')['shelf_life_days'].mean().sort_values().reset_index()
sns.barplot(data=shelf_life_cat, x='shelf_life_days', y='category', palette='OrRd_r')
plt.title('2. Average Shelf Life by Category (High Risk = Shorter Bar)', fontsize=14, fontweight='bold')
plt.xlabel('Shelf Life (Days)')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

**Chart 3: Min vs Max Stock Levels by Category**

```python
plt.figure(figsize=(10, 7))
stock_cat = df.groupby('category')[['min_stock_level', 'max_stock_level']].mean().reset_index()
stock_melted = stock_cat.melt(id_vars='category', value_vars=['min_stock_level', 'max_stock_level'],
var_name='Stock Type', value_name='Average Units')

sns.barplot(data=stock_melted, x='Average Units', y='category', hue='Stock Type', palette=['#FFA07A', '#20B2AA'])
plt.title('3. Average Stock Capacity (Min vs Max)', fontsize=14, fontweight='bold')
plt.xlabel('Number of Units')
plt.ylabel('')
plt.legend(title='Stock Type')
plt.tight_layout()
plt.show()
```

**Chart 4: Margin vs. Shelf Life Scatter (Risk vs Reward)**

```python
plt.figure(figsize=(10, 6))
prod_scatter = df.groupby('product_name').agg({
    'margin_percentage': 'mean',
    'shelf_life_days': 'mean',
    'total_price': 'sum' # Size of bubble
}).reset_index()

sns.scatterplot(data=prod_scatter, x='shelf_life_days', y='margin_percentage',
                size='total_price', sizes=(20, 800), alpha=0.6, color='purple')
plt.title('4. Risk vs Reward: Shelf Life vs Margin', fontsize=14, fontweight='bold')
plt.xlabel('Shelf Life (Days)')
plt.ylabel('Profit Margin (%)')
# Move legend safely out of the chart area
plt.legend(title='Total Revenue', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
